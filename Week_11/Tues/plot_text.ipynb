# Reading in text data from file

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

## No header information

```python
f = open("text.out","r")
n = 2**5
u = np.fromfile(f,count=n+1,sep=' ')
x = np.linspace(0,1,n+1)

plt.figure(1)
plt.plot(x,u)
```

## With header information

```python
f = open("text.out","r")
a = np.fromfile(f,count=1,sep=' ')[0]
b = np.fromfile(f,count=1,sep=' ')[0]
n = np.fromfile(f,dtype='int32',count=1,sep=' ')[0]
u = np.fromfile(f,count=n+1,sep=' ')
x = np.linspace(a,b,n+1)

plt.figure(2)
plt.plot(x,u)
```
